## üõ†Ô∏è Paso 1: Preprocesamiento - ¬°Limpiando la Casa!

Los datos del mundo real rara vez est√°n listos para usar. Necesitamos "limpiarlos" y "transformarlos". Aqu√≠ haremos:
1.  **Quitar Columnas:** Eliminaremos `mes` y `cliente` si existen, ya que no son √∫tiles como caracter√≠sticas predictivas directas en este enfoque.
2.  **Mapear Rangos:** Convertiremos los rangos de `rngEdad` y `rngSueldo` a n√∫meros representativos (como el punto medio o un valor estimado).
3.  **Convertir G√©nero:** Cambiaremos 'M' y 'F' a 1 y 0.
4.  **One-Hot Encoding:** Transformaremos variables categ√≥ricas (como `estadoCliente`, `operDigital`) en columnas num√©ricas (0/1). Esto evita que el modelo piense que 'Lima' es "m√°s grande" que 'Provincia', por ejemplo.
5.  **Manejar Nulos:** Rellenaremos cualquier valor faltante (NaN) con 0.
6.  **Convertir Booleanos:** Cambiaremos `True`/`False` (de One-Hot Encoding) a 1/0 para consistencia.

```python
# Importamos numpy para c√°lculos
import numpy as np

# Verificamos si el DataFrame 'df' del Paso 0 existe
if df is None:
  print("¬°Error! Debes ejecutar el Paso 0 y cargar tu archivo CSV primero.")
else:
  print("\nPaso 1: Limpiando y preparando tus datos...")

  # Hacemos una copia para no modificar el original
  df_processed = df.copy()

  # 1. Quitamos columnas que no usaremos (si existen)
  df_processed = df_processed.drop(['mes', 'cliente'], axis=1, errors='ignore')

  # 2. Mapas de conversi√≥n
  edad_map = {'<25-35]': 30, '<35-45]': 40, '<45-55]': 50, '>55': 60}
  sueldo_map = {'<=1000': 1000, '<1000-2500]': 1750, '<3500-4500]': 4000, '<4500-6000]': 5250, '>6000': 7000}

  # 3. Convertimos rangos a n√∫meros (de forma segura)
  if 'rngEdad' in df_processed.columns:
      df_processed['rngEdad'] = df_processed['rngEdad'].astype(str).str.strip()
      mapped_ages = df_processed['rngEdad'].map(edad_map)
      df_processed['rngEdad'] = mapped_ages.fillna(mapped_ages.median())
  if 'rngSueldo' in df_processed.columns:
      df_processed['rngSueldo'] = df_processed['rngSueldo'].astype(str).str.strip()
      mapped_sueldos = df_processed['rngSueldo'].map(sueldo_map)
      df_processed['rngSueldo'] = mapped_sueldos.fillna(mapped_sueldos.median())

  # 4. Convertimos 'genero' (si existe)
  if 'genero' in df_processed.columns:
      df_processed['genero'] = df_processed['genero'].apply(lambda x: 1 if x == 'M' else 0)

  # 5. Convertimos otras categ√≥ricas (One-Hot Encoding)
  categorical_cols = ['estadoCliente', 'procedencia', 'operDigital', 'tenTarjeta']
  # Solo intentamos convertir las columnas que realmente existen en el DataFrame
  cols_to_encode = [col for col in categorical_cols if col in df_processed.columns]
  df_processed = pd.get_dummies(df_processed, columns=cols_to_encode, drop_first=True)

  # 6. Rellenamos cualquier valor nulo que quede con 0
  df_processed = df_processed.fillna(0)

 #7. Convertir Booleanos a 1/0
  columnas_bool = df_processed.select_dtypes(include='bool').columns
  df_processed[columnas_bool] = df_processed[columnas_bool].astype(int)
  if not columnas_bool.empty:
    print(" - Columnas Booleanas (True/False) convertidas a 1/0.")

  # 8. Verificamos los tipos de datos
  print("\nAs√≠ quedaron los datos despu√©s de limpiar (Tipos):")
  df_processed.info()
  print("\nPrimeras filas de los datos procesados:")
  print(df_processed.head())

# Si 'df_processed' no se cre√≥, mostramos mensaje
if 'df_processed' not in locals():
    print("\nNo se pudo procesar el DataFrame. Revisa los pasos anteriores.")
```

## üóëÔ∏è Paso 2: Filtrar los Ceros - Enfoc√°ndonos

Nuestra variable objetivo `promSaldoPrest3Um` tiene muchos ceros. Para simplificar y mejorar la predicci√≥n del *monto*, vamos a crear un nuevo DataFrame que contenga solo los clientes que **s√≠** tienen un saldo de pr√©stamo (mayor a 0).

```python
# Verificamos si 'df_processed' del Paso 1 existe
if 'df_processed' not in locals():
  print("¬°Error! Debes ejecutar el Paso 1 para procesar los datos primero.")
else:
  print("\nPaso 2: Quitando las filas donde el monto del pr√©stamo es cero...")

  # Creamos el nuevo DataFrame filtrando
  df_no_cero = df_processed[df_processed['promSaldoPrest3Um'] > 0].copy()

  # Verificamos si quedaron datos
  if df_no_cero.empty:
    print("\n¬°Atenci√≥n! Despu√©s de quitar los ceros, no quedaron datos para analizar.")
    df_no_cero = None # Marcamos que no hay datos
  else:
    print(f"Ten√≠amos {df_processed.shape[0]} filas, ahora tenemos {df_no_cero.shape[0]} filas (solo con pr√©stamos > 0).")
    print("As√≠ se ven las primeras filas SIN CEROS:")
    print(df_no_cero.head())
```

üóëÔ∏è Paso 3: Buscamos la variable con mayot correlacion - Enfoc√°ndonos
Nuestra variable objetivo promSaldoPrest3Um , buscaremos la variable que enga mayor correlacion para un mejor entrenamiento.

```python
# Verificamos si 'df_no_cero' del Paso 2 existe
if 'df_no_cero' is None or 'df_no_cero' not in locals():
  print("¬°Error! No hay datos sin ceros para analizar (del Paso 2).")
else:
  print("\nPaso 3: Buscando la 'Mejor Pista' en los datos sin ceros...")

  # Calculamos la correlaci√≥n
  correlation_matrix_nz = df_no_cero.corr()
  correlations_with_target_nz = correlation_matrix_nz['promSaldoPrest3Um'].abs().sort_values(ascending=False)

  # La segunda es la mejor pista (la primera es ella misma)
  nombre_de_la_mejor_pista = correlations_with_target_nz.index[1]

  print("\nLas 'Mejores Pistas' (Sin Ceros):")
  print(correlations_with_target_nz)
  print(f"\n--- ¬°La 'Mejor Pista' que usaremos es: '{nombre_de_la_mejor_pista}'! ---")
```

üóëÔ∏è Paso 4:Preparamos los datos para el entrenamiento de Regresion lineal -

```python
import pandas as pd # <-- Asegurarnos que pandas est√© importado

# Verificamos si 'df_no_cero' existe
if 'df_no_cero' is None or 'df_no_cero' not in locals():
  print("¬°Error! No hay datos sin ceros para analizar. Ejecuta el Paso 2 primero.")
else:

  nombre_de_la_nueva_pista = 'promSaldoBanco3Um'  # ----------------------------- usar la pista que crees tenga mas relacion
  # -----------------------------

  print(f"\nPaso 4: Preparando X (Pista: {nombre_de_la_nueva_pista}) e y (Objetivo)...")

  # Verificamos si la nueva pista existe en el DataFrame
  if nombre_de_la_nueva_pista not in df_no_cero.columns:
      print(f"¬°Error! La columna '{nombre_de_la_nueva_pista}' no se encuentra en tus datos.")
  else:
      # X es *s√≥lo* nuestra nueva pista
      X_final_tc = df_no_cero[[nombre_de_la_nueva_pista]]

      # y sigue siendo lo que queremos adivinar
      y_final_tc = df_no_cero['promSaldoPrest3Um']

      print(f"X (Pista: {nombre_de_la_nueva_pista}) e y (Objetivo) listos.")
      print("Primeras filas de la nueva X:")
      print(X_final_tc.head())
```

üóëÔ∏è Paso 5 :Realizamos el entrenamiento del modelo de Regresion Lineal -

```python
# Importamos las librer√≠as necesarias
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np

# Verificamos si X_final_tc e y_final_tc existen
if 'X_final_tc' not in locals() or 'y_final_tc' not in locals():
  print("¬°Error! No se prepararon los datos X e y con la nueva pista (del Paso 4 Modificado).")
else:
  print(f"\nPaso 5: Entrenando y evaluando con '{nombre_de_la_nueva_pista}'...")

  # 1. Dividimos los nuevos datos
  X_train, X_test, y_train, y_test = train_test_split(X_final_tc, y_final_tc, test_size=0.3, random_state=42)

  # 2. Creamos y entrenamos el modelo
  modelo_tc = LinearRegression()
  modelo_tc.fit(X_train, y_train)

  print("\n¬°Modelo entrenado correctamente!")

  # 3. Hacemos las predicciones
  adivinanzas_tc = modelo_tc.predict(X_test)

  # 4. Evaluamos
  mae_tc = mean_absolute_error(y_test, adivinanzas_tc)
  r2_tc = r2_score(y_test, adivinanzas_tc)

  print(f"\n--- ¬°Resultados del Modelo (con {nombre_de_la_nueva_pista})! ---")
  print(f"Error Promedio (MAE): {mae_tc:.2f}")
  print(f"R-cuadrado (R^2):   {r2_tc:.2f}") # <-- ¬°Compara este R^2!

  # 5. Mostramos ejemplos
  resultados_tc = pd.DataFrame({'Monto Real': y_test, 'Adivinanza': adivinanzas_tc})
  print("\nEjemplos de Adivinanzas vs. Realidad:")
  print(resultados_tc.head())
```

üóëÔ∏è Paso 7 : Preparamos los datos para la Red Neuronal
---



```python
# Importamos las librer√≠as necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Verificamos si 'df_no_cero' existe
if 'df_no_cero' not in locals() or df_no_cero.empty:
  print("¬°Error! No hay datos sin ceros ('df_no_cero') para analizar. Ejecuta los Pasos 0, 1 y 2 primero.")
else:
  print("Paso 2: Preparando datos para la Red Neuronal...")

  # 1. Elegimos nuestras 3 variables (aseg√∫rate de que existan)
  variables_elegidas = ['rngSueldo', 'promSaldoTc3Um', 'promConsBanco3Um']
  if not all(col in df_no_cero.columns for col in variables_elegidas):
      print(f"¬°Error! No todas las variables {variables_elegidas} existen en tus datos.")
  else:
      X_nn = df_no_cero[variables_elegidas]
      y_nn = df_no_cero['promSaldoPrest3Um']

      # 2. Dividimos los datos ANTES de escalar
      X_train_nn, X_test_nn, y_train_nn, y_test_nn = train_test_split(X_nn, y_nn, test_size=0.2, random_state=42) # Usamos 20% para prueba

      # 3. Escalamos los datos (MUY importante para redes neuronales)
      scaler_x = StandardScaler()
      X_train_scaled = scaler_x.fit_transform(X_train_nn)
      X_test_scaled = scaler_x.transform(X_test_nn)

      # Es buena idea escalar tambi√©n 'y', pero a veces complica la interpretaci√≥n.
      # Por ahora, la dejaremos como est√°, pero tenlo en cuenta.

      print("¬°Datos listos y escalados!")
      print(f"Forma de X_train_scaled: {X_train_scaled.shape}")
      print(f"Forma de X_test_scaled: {X_test_scaled.shape}")
```

Paso 8
: Construyendo la Red Neuronal...

```python
# Importamos las librer√≠as de Keras (TensorFlow)
import tensorflow as tf
from tensorflow import keras
from keras import layers

# Verificamos si tenemos los datos listos
if 'X_train_scaled' not in locals():
    print("¬°Error! Los datos no est√°n preparados (Paso 2).")
else:
    print("\nPaso 3: Construyendo la Red Neuronal...")

    # Definimos el modelo como una secuencia de capas
    modelo_nn = keras.Sequential()

    # Capa de Entrada y Primera Capa Oculta:
    # - input_shape debe coincidir con el n√∫mero de variables (3)
    # - 16 neuronas (puedes probar con otros n√∫meros)
    # - 'relu' es una funci√≥n de activaci√≥n muy com√∫n y buena para empezar.
    modelo_nn.add(layers.Dense(16, input_shape=(X_train_scaled.shape[1],), activation='relu'))

    # Segunda Capa Oculta (Opcional, pero la a√±adimos)
    modelo_nn.add(layers.Dense(8, activation='relu'))

    # Capa de Salida:
    # - 1 neurona porque queremos predecir UN valor (el monto del pr√©stamo).
    # - 'linear' es la activaci√≥n est√°ndar para regresi√≥n (predecir n√∫meros).
    modelo_nn.add(layers.Dense(1, activation='linear'))

    # Compilamos el modelo:
    # - optimizer='adam' es un buen optimizador general.
    # - loss='mae' (Mean Absolute Error) es la m√©trica que intentaremos minimizar.
    # - metrics=['mae', 'mse'] nos mostrar√° MAE y MSE durante el entrenamiento.
    modelo_nn.compile(optimizer='adam', loss='mae', metrics=['mae', 'mse'])

    # Mostramos un resumen de c√≥mo qued√≥ nuestra red
    print("\n¬°Red Neuronal Construida! Aqu√≠ est√° el resumen:")
    modelo_nn.summary()
```

Paso 9
: Entrenando la Red Neuronal (¬°Paciencia!)...


```python
# Verificamos si tenemos el modelo y los datos
if 'modelo_nn' not in locals() or 'X_train_scaled' not in locals():
    print("¬°Error! El modelo o los datos no est√°n listos (Pasos 2 y 3).")
else:
    print("\nPaso 4: Entrenando la Red Neuronal (¬°Paciencia!)...")

    # Entrenamos el modelo
    # - epochs=100 significa que pasaremos por todos los datos 100 veces.
    # - validation_split=0.2 usa el 20% de los datos de entrenamiento para validar
    #   en cada √©poca, ayud√°ndonos a ver si sobreajusta.
    # - batch_size=32 procesa los datos en grupos de 32.
    # - verbose=1 nos muestra el progreso.
    history = modelo_nn.fit(
        X_train_scaled,
        y_train_nn,
        epochs=100,
        validation_split=0.2,
        batch_size=32,
        verbose=1
    )

    print("\n¬°Entrenamiento Completado!")
```

Paso 10: Graficando la evoluci√≥n del error...



```python
import matplotlib.pyplot as plt

# Verificamos si tenemos el historial
if 'history' not in locals():
    print("¬°Error! El modelo no ha sido entrenado (Paso 4).")
else:
    print("\nPaso 5: Graficando la evoluci√≥n del error...")

    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure(figsize=(10, 6))
    plt.xlabel('Epoch (Vueltas de Entrenamiento)')
    plt.ylabel('Mean Abs Error (MAE) - Error Promedio')
    # Graficamos el MAE en los datos de entrenamiento
    plt.plot(hist['epoch'], hist['mae'], label='Error de Entrenamiento')
    # Graficamos el MAE en los datos de validaci√≥n
    plt.plot(hist['epoch'], hist['val_mae'], label='Error de Validaci√≥n')
    # Puedes ajustar el l√≠mite del eje Y si el error inicial es muy alto
    # plt.ylim([0, 50000]) # Descomenta y ajusta si es necesario
    plt.legend()
    plt.grid(True)
    plt.title('Evoluci√≥n del Error (MAE) Durante el Entrenamiento')
    plt.show()
```

Paso 11: Evaluando y Graficando Predicciones...


```python
# Verificamos si tenemos el modelo y los datos de prueba
if 'modelo_nn' not in locals() or 'X_test_scaled' not in locals():
    print("¬°Error! El modelo o los datos de prueba no est√°n listos.")
else:
    print("\nPaso 6: Evaluando y Graficando Predicciones...")

    # Hacemos las predicciones con los datos de prueba (escalados)
    predicciones_nn = modelo_nn.predict(X_test_scaled).flatten() # .flatten() convierte la salida a un array simple

    # Creamos un gr√°fico de dispersi√≥n para comparar
    plt.figure(figsize=(10, 8))
    plt.scatter(y_test_nn, predicciones_nn, alpha=0.5)
    plt.xlabel('Valores Reales (Monto Pr√©stamo)')
    plt.ylabel('Predicciones del Modelo')
    plt.title('Valores Reales vs. Predicciones de la Red Neuronal')

    # A√±adimos una l√≠nea diagonal (y=x). Si los puntos est√°n cerca de esta
    # l√≠nea, significa que las predicciones son buenas.
    lim_min = min(y_test_nn.min(), predicciones_nn.min())
    lim_max = max(y_test_nn.max(), predicciones_nn.max())
    plt.plot([lim_min, lim_max], [lim_min, lim_max], 'r--', lw=2, label='Predicci√≥n Perfecta')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculamos el R^2 para la red neuronal
    from sklearn.metrics import r2_score
    r2_nn = r2_score(y_test_nn, predicciones_nn)
    print(f"\nCoeficiente de Determinaci√≥n (R^2) de la Red Neuronal: {r2_nn:.2f}")
```

--- An√°lisis Adicional con L√≥gica de Programaci√≥n ---


```python
import matplotlib.pyplot as plt
import numpy as np # Asegurarnos que numpy est√© importado

# Verificamos si tenemos las predicciones y los valores reales
if 'predicciones_nn' not in locals() or 'y_test_nn' not in locals():
    print("¬°Error! No se han generado las predicciones (Paso 6 anterior).")
else:
    print("\n--- An√°lisis Adicional con L√≥gica de Programaci√≥n ---")

    # 1. Definimos umbrales y un DICCIONARIO para contar
    #    Estos umbrales son ejemplos, ¬°puedes ajustarlos!
    umbrales = {
        'Bajo': 5000,
        'Medio': 20000,
        'Alto': np.inf # Usamos infinito para capturar todo lo dem√°s
    }

    # Creamos un diccionario para guardar nuestros conteos, inicializado en 0
    conteo_categorias = {'Bajo': 0, 'Medio': 0, 'Alto': 0}

    # 2. Creamos una LISTA para guardar las diferencias (errores)
    errores_prediccion = []

    # 3. Usamos un BUCLE (for) para recorrer cada predicci√≥n y valor real
    #    'zip' nos permite recorrer y_test_nn y predicciones_nn al mismo tiempo
    for real, predicho in zip(y_test_nn, predicciones_nn):

        # Guardamos el error absoluto en nuestra lista
        errores_prediccion.append(abs(real - predicho))

        # 4. Usamos CONDICIONALES (if/elif/else) para clasificar el valor REAL
        if real < umbrales['Bajo']:
            conteo_categorias['Bajo'] += 1
        elif real < umbrales['Medio']:
            conteo_categorias['Medio'] += 1
        else:
            conteo_categorias['Alto'] += 1

    # 5. Mostramos los resultados
    print("\nConteo de Pr√©stamos Reales por Categor√≠a:")
    # Usamos un bucle para mostrar el diccionario
    for categoria, cantidad in conteo_categorias.items():
        print(f"- {categoria}: {cantidad} pr√©stamos")

    # Calculamos el error promedio desde nuestra lista
    error_promedio_calculado = sum(errores_prediccion) / len(errores_prediccion)
    print(f"\nError Absoluto Medio (calculado manualmente): {error_promedio_calculado:.2f}")

    # 6. Visualizaci√≥n extra (Histograma de Errores)
    plt.figure(figsize=(10, 6))
    plt.hist(errores_prediccion, bins=50, alpha=0.7, color='orange')
    plt.title('Distribuci√≥n de los Errores de Predicci√≥n')
    plt.xlabel('Error Absoluto (Real - Predicho)')
    plt.ylabel('Frecuencia (Cu√°ntas veces ocurre)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```
